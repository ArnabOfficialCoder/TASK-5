```python
# Step 1: Import required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Make plots look nice
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

```

```python
# Step 2: Load CSV files
train_df = pd.read_csv(r"A:\ELEVATE GROUP\Data Analyst Internship\Day 5\archive (1)\train.csv")
test_df = pd.read_csv(r"A:\ELEVATE GROUP\Data Analyst Internship\Day 5\archive (1)\test.csv")


```

```python
# Step 3: Quick peek at the data
train_df.head()
```

```python
# Info about the dataset
train_df.info()


```

```python
# Summary statistics for numeric columns
train_df.describe()
```

```python

# Missing values count
train_df.isnull().sum()
```

```python
# Count of passengers by gender
train_df['Sex'].value_counts()
```

```python
# Passenger class counts
train_df['Pclass'].value_counts()
```

```python
# Embarked port counts
train_df['Embarked'].value_counts()
```

```python
# Survival rate by gender
train_df.groupby('Sex')['Survived'].mean()
```

```python
# Survival rate by class
train_df.groupby('Pclass')['Survived'].mean()
```

```python
# Survival rate by embarkation port
train_df.groupby('Embarked')['Survived'].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Pairplot - good for continuous variables
sns.pairplot(train_df, diag_kind="kde")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(train_df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.barplot(x="Sex", y="Survived", data=train_df)
plt.title("Survival Rate by Gender")
plt.show()
```

```python
sns.barplot(x="Pclass", y="Survived", data=train_df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
train_df.hist(figsize=(12, 8), bins=20)
plt.suptitle("Feature Distributions")
plt.show()
```

```python
sns.boxplot(x="Pclass", y="Age", data=train_df)
plt.title("Age Distribution by Passenger Class")
plt.show()
```

```python
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=train_df)
plt.title("Age vs Fare (Colored by Survival)")
plt.show()
```
